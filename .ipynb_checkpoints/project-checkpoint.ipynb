```python
heart_data['healthy_foods'] = heart_data['Fruits'] + heart_data['Veggies'] #np.where( ((heart_data['Fruits'] == 1) | (heart_data['Veggies'] == 1)), 1, 0)
```

```python
heart_data['healthcare_difficulty'] = np.where(((heart_data['NoDocbcCost'] == 1) | (heart_data['AnyHealthcare'] == 0)), 1, 0)
```

```python
heart_data = heart_data.drop(['NoDocbcCost', 'AnyHealthcare', 'Fruits', 'Veggies'], axis = 1)
```

```python
heart_data.iloc[:,15:]
```

```python
plt.hist(heart_data[['PhysHlth']])
```

```python
heart_data['PhysHlth']
```

```python
heart_data['phys_health_group'] = 0
intervals = sorted(pd.unique(pd.cut(heart_data['PhysHlth'], bins = 5)))
for i in range(0, len(intervals)):
    mask = heart_data['PhysHlth'].between(intervals[i].left, intervals[i].right, inclusive = 'right')
    heart_data['phys_health_group'] = np.where(mask == True, i + 1, heart_data['phys_health_group'])
```

```python
heart_data['phys_health_group'].value_counts()
```

```python
plt.hist(heart_data[['MentHlth']])
```

```python
intervals = sorted(pd.unique(pd.cut(heart_data['MentHlth'], bins = 5)))
intervals
```

```python
heart_data['ment_health_group'] = 0
intervals = sorted(pd.unique(pd.cut(heart_data['MentHlth'], bins = 5)))
for i in range(0, len(intervals)):
    mask = heart_data['MentHlth'].between(intervals[i].left, intervals[i].right, inclusive = 'right')
    heart_data['ment_health_group'] = np.where(mask == True, i + 1, heart_data['ment_health_group'])
```

```python
heart_data['ment_health_group'].value_counts()
```

```python
heart_data = heart_data.drop(['PhysHlth', 'MentHlth'], axis = 1)
```

```python
heart_data
```

```python
cor = heart_data.corr()
```

```python
from imblearn.over_sampling import SMOTE
# import imblearn
```

```python
X = heart_data.iloc[:, 1:]
y = heart_data.iloc[:, 0]
print(X.shape)
print(y.shape)
```

```python
oversample = SMOTE()
X_new, y_new = oversample.fit_resample(X, y)
```

```python
oversampled_data = X_new.join(y_new)
oversampled_data.head()
```

```python
fig = plt.figure(figsize = (12, 10))
oversampled_cor = oversampled_data.corr()
sns.heatmap(oversampled_cor, cmap = 'coolwarm')
```

```python
cor_diff = cor - oversampled_cor
cor_diff[abs(cor_diff) > 0.1]
```

```python
cor[cor['HeartDiseaseorAttack'] < 0].index
```

```python
oversampled_cor[oversampled_cor['HeartDiseaseorAttack'] < 0].index
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y_new, test_size = 0.3)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)
```

```python
rfc_scores = cross_val_score(rfc, X_new, y_new, cv = 5)
```

```python
rfc_cv_score = np.mean(rfc_scores)
rfc_cv_score
```

```python
lr = LogisticRegression(max_iter = 250)
lr_scores = cross_val_score(lr, X_new, y_new, cv = 5)
```

```python
lr_cv_score = np.mean(lr_scores)
lr_cv_score
```

```python
dtc = DecisionTreeClassifier()
dtc_scores = cross_val_score(dtc, X_new, y_new, cv = 5)
```

```python
dtc_cv_score = np.mean(dtc_scores)
dtc_cv_score
```

```python
svc = LinearSVC()
svc_scores = cross_val_score(svc, X_new, y_new, cv = 5)
```

```python
svc_cv_score = np.mean(svc_scores)
svc_cv_score
```

```python
gnb = GaussianNB()
gnb_scores = cross_val_score(gnb, X_new, y_new, cv = 5)
```

```python
gnb_cv_score = np.mean(gnb_scores)
gnb_cv_score
```

```python
model_names = ['Random Forest', 'Logistic Regression', 'Decision Tree', 'LinearSVC', 'Naive Bayes']
model_scores = [rfc_cv_score, lr_cv_score, dtc_cv_score, svc_cv_score, gnb_cv_score]
for i in range(0, len(model_names)):
    print(model_names[i], 'Cross Validation Score:', round(model_scores[i],3))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
```

Random Forest performs the best for the oversampled data. Let's see what happens when we undersample.

```python
from imblearn.under_sampling import NearMiss
```

```python
undersample = NearMiss(version = 3)
```

```python
X_under, y_under = undersample.fit_resample(X, y)
```

```python
print(X_under.shape)
print(y_under.shape)
```

```python
undersampled_data = X_under.join(y_under)
undersampled_data.head()
```

```python
fig = plt.figure(figsize = (12, 10))
undersampled_cor = undersampled_data.corr()
sns.heatmap(undersampled_cor, cmap = 'coolwarm')
```

```python
under_cor_diff = cor - undersampled_cor
under_cor_diff[abs(under_cor_diff) > 0.1]
```

```python
cor[cor['HeartDiseaseorAttack'] < 0].index
```

```python
undersampled_cor[undersampled_cor['HeartDiseaseorAttack'] < 0].index
```

Undersampling significantly changes the way features interact with our target, likely we will need to use oversampling instead as oversampling preserves the way features affect the target better.

```python
X_train, X_test, y_train, y_test = train_test_split(X_under, y_under, test_size = 0.3)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
rfc = RandomForestClassifier()
rfc_scores = cross_val_score(rfc, X_under, y_under, cv = 5)
```

```python
rfc_cv_score = np.mean(rfc_scores)
rfc_cv_score
```

```python
lr = LogisticRegression(max_iter = 500)
lr_scores = cross_val_score(lr, X_under, y_under, cv = 5)
```

```python
lr_cv_score = np.mean(lr_scores)
lr_cv_score
```

```python
dtc = DecisionTreeClassifier()
dtc_scores = cross_val_score(dtc, X_under, y_under, cv = 5)
```

```python
dtc_cv_score = np.mean(dtc_scores)
dtc_cv_score
```

```python
svc = LinearSVC(max_iter = 2000)
svc_scores = cross_val_score(svc, X_under, y_under, cv = 5)
```

```python
svc_cv_score = np.mean(svc_scores)
svc_cv_score
```

```python
gnb = GaussianNB()
gnb_scores = cross_val_score(gnb, X_under, y_under, cv = 5)
```

```python
gnb_cv_score = np.mean(gnb_scores)
gnb_cv_score
```

```python
model_names = ['Random Forest', 'Logistic Regression', 'Decision Tree', 'LinearSVC', 'Naive Bayes']
model_scores = [rfc_cv_score, lr_cv_score, dtc_cv_score, svc_cv_score, gnb_cv_score]
for i in range(0, len(model_names)):
    print(model_names[i], 'Cross Validation Score:', round(model_scores[i],3))
```

The best cross validation score we could achieve using undersampling was 0.648 with logistic regression. It looks like we will need to use oversampling as a more reliable method.
